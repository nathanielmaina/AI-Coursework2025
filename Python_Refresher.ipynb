```python
# Question 2 Calculating volume of sphere
radius= float(input( " Enter the radius in order to calculate volume of the sphere (in metres): "))
cubedR= radius**3  #use of expotential operator
volume = 4/3 * 3.142* cubedR
print( "The volume of sphere is :",volume)
```

```python
# Question 3 Calculate perimeter and area of a square
sides=float(input( "Enter length of square (in metres):"))

def perimeter(sides):
    return 4*sides

def area(sides):
    return sides*sides

print("Perimeter of the square is", perimeter(sides),"and area of square is",area(sides))
```

```python
#Question 6  use of loop to generate an average value

numbers=[]

for i in range(5):
    num = float(input("Enter your values{i+1}: "))
    numbers.append(num)

average = sum(numbers) / len(numbers)

print("The average value is",average)

```
